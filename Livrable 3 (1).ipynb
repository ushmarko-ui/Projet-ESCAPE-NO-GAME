# II. Impact de la chaîne de transmission 

## A) Transmission d'un texte court :

* Taille faible → Moins de données à transmettre.
* Faible besoin en bande passante → Transmission rapide.
* Encodage simple → ASCII ou Unicode.
* Faible impact sur la chaîne → Peut être envoyé sous forme de paquets légers avec peu de risques d’erreur.

Source (Texte écrit) → Encodage (ASCII ou Unicode) → Protocole de transmission (HTTP ou SMTP) → Réseau (Internet ou intranet) → Réception (Destination)

## B) Transmission d’un fichier texte volumineux : 

* Taille importante → Plus de paquets à transmettre.
* Impact sur la bande passante → Nécessite un débit plus élevé.
* Risque de fragmentation → Nécessite un protocole de correction d’erreur et d’optimisation (ex. compression).
* Possible latence → Surtout si le réseau est encombré.

Source (Fichier texte) → Compression (ZIP, GZIP) → Encodage (UTF-8 ou autre) → Protocole de transmission (TCP pour la fiabilité) → Réseau (Internet à large bande passante) → Réception (Destination) → Décompression (ZIP)

## C) Transmission d’un fichier binaire quelconque (ex. image, vidéo, programme)

* Taille variable, souvent grande → Débit élevé requis.
* Encodage complexe → Dépend du type de fichier (JPEG, MP3, MP4, EXE...).
* Nécessite des protocoles de fiabilité → TCP pour éviter les pertes de données.
* Peut nécessiter une compression → ZIP, RAR, formats optimisés.

Source (Image/vidéo) → Compression (JPEG, MP4, PNG) → Encodage (selon format) → Protocole de transmission (TCP pour éviter la perte de données) → Réseau (large bande passante, fibre optique) → Réception (Destination) → Décompression/affichage (dépend du type de fichier)

## D) Transmission d’un son (ex. Voix, Musique)

* Données volumineuses si non compressées → WAV non compressé vs MP3 compressé.
* Bande passante critique → Nécessite un débit constant pour éviter les coupures.
* Nécessité de codecs → MP3, AAC, FLAC pour optimiser la taille et la qualité.
* Protocole en temps réel requis → UDP (faible latence) ou TCP (meilleure fiabilité).

Source (Voix/Musique) → Encodage (MP3, AAC) → Protocole de transmission (UDP pour faible latence ou RTP pour audio en temps réel) → Réseau (Internet ou réseau local avec faible latence) → Réception (Destination) → Décodage (MP3, AAC) → Lecture (haut-parleur ou casque)



# III. Détail des étapes de la chaîne de transmission 

## A) Message 

L’information initiale est capturée sous forme de signal sonore, qui doit être converti en données numériques pour être transmis. Cette étape repose sur un microphone, qui capte les vibrations sonores et les transforme en un signal électrique analogique.Le microphone filaire a été privilégié en raison de sa fiabilité et de sa faible sensibilité aux perturbations extérieures.

## B) Conversion ASCII & Binaire





Une fois le signal capté, il doit être numérisé pour pouvoir être traité par un ordinateur. Cette numérisation s’effectue en plusieurs étapes :

1.	Échantillonnage : Le signal est prélevé à intervalles réguliers.
2.	Quantification : Chaque échantillon est converti en une valeur numérique.
3.	Encodage : Ces valeurs sont représentées sous forme binaire.

L’encodage en ASCII (8 bits par caractère) a été retenu car il est simple et efficace pour des transmissions textuelles. Pour un signal vocal, un encodage en binaire brut avec compression est préférable.

## C) Codage en bande de base 

Avant d’être transmis, le signal numérique doit être préparé pour minimiser les erreurs et optimiser la transmission
Le codage Manchester a été choisi pour assurer une synchronisation optimale du signal, bien qu’il double la bande passante nécessaire.

## D) Modulation 



La modulation permet d’adapter le signal numérique à un support de transmission (câble ou ondes).
La modulation FSK (Frequency Shift Keying) a été retenue, car elle offre une meilleure résistance aux interférences.

## E) Transmission 

Une fois modulé, le signal est envoyé à travers un canal de communication.
Les ultrasons sont utilisées lorsque la mobilité est nécessaire.

## F) Réception 

Le signal est capté par le dispositif récepteur, qui doit l’interpréter correctement
L’ordinateur sur écoute utilise une carte réseau filaire ou sans fil en fonction du mode de transmission initial.

## G) Démodulation 

Le signal reçu est reconverti en données exploitables en effectuant l’opération inverse de la modulation.
Le démodulateur FSK est utilisé pour être cohérent avec la modulation choisie à l’émission.

## H) Décodage binaire et ASCII 

Le signal numérique est transformé en données lisibles.
L’encodage ASCII est utilisé pour les textes, tandis qu’un décodage spécifique est appliqué pour les fichiers binaires.

## I) Message restitué 

Enfin, les données sont converties en un format compréhensible par l’utilisateur.
Le message peut être affiché sous forme de texte à l’écran ou restitué vocalement par les haut-parleurs.

# IV. Quantification au maximum de toutes les données :

Un texte court typique (par exemple, une phrase de 100 caractères en ASCII)occupe généralement entre 100 et 200 octets, selon l'encodage.

Estimation :

Nombre de caractères : 100 (en ASCII, 1 octet par caractère)
Taille totale : 100 octets

Calcul pour un transfert :

Bande passante nécessaire : Faible (quelques octets par seconde suffisent, donc < 1 kbps).

## B) Transmission d'un fichier texte volumineux 

Un fichier texte de taille 50 Mo. Ce fichier sera envoyé via TCP (pour assurer la fiabilité), et nous pouvons également appliquer une compression de 50%.

Estimation :

Taille initiale du fichier : 50 Mo = 50 * 1024 * 1024 octets = 52 428 800 octets
Compression (50%) : 52 428 800 octets * 0.5 = 26 214 400 octets
Transmission (avec TCP) :

Bande passante nécessaire : Si le fichier est transféré sur 10 secondes, la bande passante nécessaire serait de :

26214400 octets/ 10 secondes = 2621440 octets/s=21Mbps

## C) Transmission d’un fichier binaire (ex. image, vidéo, programme)

Prenons un fichier vidéo de 500 Mo en format MP4.

Estimation :

Taille du fichier vidéo : 500 Mo = 500 * 1024 * 1024 octets = 524 288 000 octets
Compression : MP4 est déjà un format compressé, donc considérons qu'il n'y a pas de compression supplémentaire (ce qui est souvent le cas).

Transmission (avec TCP) :

Bande passante nécessaire : Si le fichier doit être transféré en 30 secondes, la bande passante nécessaire serait :

524288000 octets / 30 secondes =17476267 octets/s=139Mbps

# VII. Exemple concret 

## A) Petit intro en ASCII : 

Chaque caractère a une valeur en ASCII (American Standard Code for Information Interchange). Cette valeur est ensuite convertie en binaire. 

Exemple : Prenons la lettre "A" : 

Sa valeur ASCII est 65. 

Sa représentation binaire sur 8 bits est 01000001. 

Méthode : 

Trouver la valeur ASCII du caractère (ex. "A" = 65). 

Convertir ce nombre en binaire (65 → 01000001). 

Répéter pour chaque caractère du texte. 

Exemple avec "Hello" : 

H → 72 → 01001000 

e → 101 → 01100101 

l → 108 → 01101100 

l → 108 → 01101100 

o → 111 → 01101111 

## B) Encodage (préparation du message)

Le caractère "A" en ASCII a une valeur binaire de 01000001 (en binaire sur 8 bits).

* Type de donnée : Caractère ASCII
* Signal d'entrée : Le caractère "A"
* Transformation : Le caractère est converti en sa représentation binaire.

-- Représentation binaire :

* "A" = 01000001 (en binaire sur 8 bits)
* Taille du message : 1 octet

## C) Transmission (encapsulation dans un paquet de données)

Une fois que le caractère est encodé en binaire, il est préparé pour la transmission. Si nous utilisons un protocole de transmission comme TCP, le message est encapsulé dans un paquet.

* Type de donnée : Paquet TCP contenant des octets

-- Signal d'entrée :

* Paquet TCP contenant l'octet 01000001
* Le paquet TCP inclura également un en-tête avec des informations comme le numéro de séquence et les informations de contrôle.

-- Transformation :

* Le caractère est envoyé en tant que paquet TCP qui contient à la fois les données utiles (l'octet) et les informations de gestion du protocole (numéro de séquence, somme de contrôle, etc.).
* Représentation dans le paquet TCP :
* Paquet TCP → [En-tête TCP] + [Données : 01000001]
* Taille totale du paquet TCP : Varie selon l'en-tête TCP, mais les données utiles restent 1 octet.

## D) Transmision via le réseau 

Le paquet est envoyé à travers le réseau. À ce stade, le paquet est peut-être fragmenté selon le type de réseau utilisé (Ethernet, Wi-Fi, etc.), mais pour simplifier, nous allons supposer que le paquet arrive intact.

* Type de donnée : Paquet de données
* Signal d'entrée : Le paquet TCP contenant l'octet 01000001
* Transformation : Le paquet traverse le réseau (câblé ou sans fil) jusqu'à la destination.

## E) Réception du paquet

Le récepteur reçoit le paquet TCP contenant l'octet 01000001.

* Type de donnée : Paquet TCP contenant des données binaires
* Signal d'entrée : Paquet TCP reçu contenant 01000001 (et éventuellement des informations de contrôle, comme le numéro de séquence, mais ce n'est pas nécessaire pour notre exemple simple).
* Transformation : Le récepteur analyse le paquet TCP, vérifie l'intégrité et extrait les données utiles (ici, l'octet 01000001).

## F) Décodage du message 

Le récepteur va maintenant convertir l'octet binaire 01000001 en son caractère ASCII correspondant.

* Type de donnée : Caractère ASCII
* Signal d'entrée : Octet binaire 01000001

-- Transformation :

* Conversion de l'octet binaire 01000001 en son caractère ASCII associé, qui est "A".
* Représentation en texte : Le caractère "A" est maintenant obtenu.

## G) Affichage du message 

Le caractère "A" est maintenant prêt à être utilisé, affiché à l'écran, ou traité par l'application du récepteur.

* Type de donnée : Caractère ASCII
* Signal de sortie : Le caractère "A" est affiché à l'écran ou utilisé dans le programme du récepteur.
